```python
from keras.preprocessing import image
from keras.models import load_model
import numpy as np
import os
```

```python
#We first load our image
def load_image(image_path):
  loaded_img = image.load_img(image_path,target_size=(256,256),color_mode ='grayscale')
  image_tensor = image.img_to_array(loaded_img)
  image_tensor = np.expand_dims(image_tensor,axis=0)
  image_tensor /= 255.
  return image_tensor



```

```python
model = load_model("./drive/MyDrive/Dataset.prediction.h5")

tn=tp=fn=fp =0
```

```python
for img in os.listdir("./drive/MyDrive/Dataset/test/yawn"):
  predictions = model.predict(load_image("./drive/MyDrive/Dataset/test/yawn/" + img))
  if predictions[0][0] < predictions[0][1]:
    tp+=1
  else:
    fn+=1

for img in os.listdir("./drive/MyDrive/Dataset/test/no_yawn"):
    predictions = model.predict(load_image("./drive/MyDrive/Dataset/test/no_yawn/" + img))
    if predictions[0][0] > predictions[0][1]:  # predicted 'not yawn'
        tn += 1
    else:  # predicted 'yawn'
        fp += 1

# calculate precision, recall and f1_score
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1_score = 2 * precision * recall / (precision + recall)

# save results
open("./drive/MyDrive/confusion_matrix.txt", "w").write("TP: " + str(tp) + "\nTN: " + str(tn) + "\nFP: " + str(fp) + "\nFN: " + str(fn) + "\n\nPrecision: " + str(precision) + "\nRecall: " + str(recall) + "\nF1 score: " + str(f1_score))
```

```python
print(precision)
print(f1_score)
print(recall)
```
